```python
import os

dizin_yolu = '/content/runs/detect/predict3/TurnikeKısaINPUT_frames'

# Dizin içindeki tüm dosyaları al
dosyalar = os.listdir(dizin_yolu)

for dosya in dosyalar:
    # Yeni dosya adını oluştur
    yeni_ad = dosya.replace('turnıkeIn_', '')

    # Dosyanın tam yolu
    eski_yol = os.path.join(dizin_yolu, dosya)
    yeni_yol = os.path.join(dizin_yolu, yeni_ad)

    # Dosyayı yeniden adlandır
    os.rename(eski_yol, yeni_yol)

print("Dosya adları başarıyla değiştirildi.")
```

```python
import os

# Dizin yolu
directory = '/content/runs/detect/predict/labels/'

# Dizindeki dosyaları listele
files = os.listdir(directory)

# Dosyalar üzerinde dön
for file in files:
    # Dosyanın tam yolu
    file_path = os.path.join(directory, file)

    # Dosyanın adını al (örneğin: onlukIn_102.txt)
    file_name = os.path.basename(file_path)

    # Önek ve son ek kısmını kaldır
    new_file_name = file_name.replace('onlukIn_', '').replace('.txt', '')

    # Yeni dosya adı
    new_file_name = new_file_name + '.txt'

    # Dosyanın yeni tam yolu
    new_file_path = os.path.join(directory, new_file_name)

    # Dosyanın adını değiştir
    os.rename(file_path, new_file_path)

print("Dosya adı düzenleme işlemi tamamlandı.")
```

```python
import os

# Dizin ve hedef etiket adını belirtin
dizin = "/content/runs/detect/predict2/onlukIn_frames/"
hedef_etiket = "onlukln_"

# Dizindeki tüm dosya adlarını al
dosya_adlari = os.listdir(dizin)

# Her dosya için yeni adı oluştur ve dosyayı yeniden adlandır
for dosya_adı in dosya_adlari:
    eski_yol = os.path.join(dizin, dosya_adı)
    yeni_ad = dosya_adı.split('.')[0]  # Dosya uzantısını kaldır
    yeni_ad = hedef_etiket + yeni_ad + '.jpg'  # Yeni etiketle ve uzantıyı ekle
    yeni_yol = os.path.join(dizin, yeni_ad)
    os.rename(eski_yol, yeni_yol)

print("Dosya adları başarıyla değiştirildi.")
```

```python
import os
import shutil

# Kaynak ve hedef dizinleri belirtin
kaynak_dizin = "/content/runs/detect/predict2/labels/"
hedef_dizin = "/content/xx/labels/"

# Hedef dizini oluştur (eğer henüz yoksa)
os.makedirs(hedef_dizin, exist_ok=True)

# Kaynak dizinindeki tüm dosyaları hedef dizine taşı
for dosya_adı in os.listdir(kaynak_dizin):
    kaynak_yol = os.path.join(kaynak_dizin, dosya_adı)
    hedef_yol = os.path.join(hedef_dizin, dosya_adı)
    shutil.move(kaynak_yol, hedef_yol)

print("Dosyalar başarıyla taşındı.")
```
