```python
import os

# Dizin yolu
directory = '/content/runs/detect/predict/labels/'

# Dizindeki dosyalarÄ± listele
files = os.listdir(directory)

# Dosyalar Ã¼zerinde dÃ¶n
for file in files:
    # DosyanÄ±n tam yolu
    file_path = os.path.join(directory, file)

    # DosyanÄ±n adÄ±nÄ± al (Ã¶rneÄŸin: onlukIn_102.txt)
    file_name = os.path.basename(file_path)

    # Ã–nek ve son ek kÄ±smÄ±nÄ± kaldÄ±r
    new_file_name = file_name.replace('onlukIn_', '').replace('.txt', '')

    # Yeni dosya adÄ±
    new_file_name = new_file_name + '.txt'

    # DosyanÄ±n yeni tam yolu
    new_file_path = os.path.join(directory, new_file_name)

    # DosyanÄ±n adÄ±nÄ± deÄŸiÅŸtir
    os.rename(file_path, new_file_path)

print("Dosya adÄ± dÃ¼zenleme iÅŸlemi tamamlandÄ±.")
```
