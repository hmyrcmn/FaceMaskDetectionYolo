```python
import cv2
import torch
from tqdm.notebook import tqdm

video_path = '/content/MAKK-HAIRNET-INPUT.mp4'
cap = cv2.VideoCapture(video_path)

# YOLO modelini yükleyin
model =YOLO(/content/Result2/runs/detect/train2/weights/best.pt)

# YOLO modelinin sınıf listesini alın
class_names = model.names

# Tahmin eşik değeri
threshold = 0.6

# Çıkış video dosyasını oluşturun
output_path = '/content/output_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out = cv2.VideoWriter(output_path, fourcc, 30, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # OpenCV ile görüntüyü RGB formatına dönüştürün
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # YOLO modelini kullanarak nesneleri tespit edin
    results = model(frame_rgb)

    # Tahmin değeri 0.6'dan büyük olan nesneleri filtreleyin
    filtered_results = results.pred[0][results.pred[0][:, 4] >= threshold]

    # Filtrelenmiş sonuçları çerçeve üzerine çizin
    frame = model.plot([filtered_results], img=frame_rgb, names=class_names, colors=model.colors(2))

    # Sonucu video dosyasına yazın
    out.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))

cap.release()
out.release()
cv2.destroyAllWindows()
```
