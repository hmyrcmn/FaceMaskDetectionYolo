```python
import cv2

# Ä°ki video dosyasÄ±nÄ± aÃ§Ä±n
video1 = cv2.VideoCapture('/content/runs/detect/predict3/hairnetinput3.avi')
video2 = cv2.VideoCapture('/content/runs/detect/predict4/hairnetinput3.avi')

# Ä°lk videonun Ã¶zelliklerini alÄ±n
frame_width = int(video1.get(3))
frame_height = int(video1.get(4))
frame_rate = int(video1.get(5))

# Ã‡Ä±kÄ±ÅŸ video dosyasÄ±nÄ± oluÅŸturun
output_video = cv2.VideoWriter('combined_video.avi', cv2.VideoWriter_fourcc(*'XVID'), frame_rate, (frame_width, frame_height))

while True:
    ret1, frame1 = video1.read()
    ret2, frame2 = video2.read()

    if not ret1 or not ret2:
        break

    # Ä°ki Ã§erÃ§eveyi yatay olarak birleÅŸtirin
    combined_frame = cv2.hconcat([frame1, frame2])

    # BirleÅŸtirilmiÅŸ Ã§erÃ§eveyi Ã§Ä±kÄ±ÅŸ video dosyasÄ±na yazÄ±n
    output_video.write(combined_frame)

# Video dosyalarÄ±nÄ± kapatÄ±n
video1.release()
video2.release()
output_video.release()
```
